```python
df1 = df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date','Type', 
       'Manufacturer', 
       # 'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 'Opened', 'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 
       'latitude', 
       'longitude', 
       'Type_Main',
       'opening_date_clean', 
       #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
       #'height_value', 'height_unit', 
       'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

2.2 Change the data type of the column

```python
df1['opening_date_clean'] = pd.to_datetime(df1['opening_date_clean'])
```

2.3 Rename the column names


```python
df1.columns
```

```python
df2 = df1.rename(columns={'coaster_name': 'Coaster_Name',
                    'year_introduced': 'Year_Introduced',
                    'latitude': 'Latitude',
                    'longitude': 'Longitude',
                    'opening_date_clean': 'Opening_Date',
                    'speed_mph': 'Speed-Mph',
                    'height_ft': 'Height-Ft'})
```

```python
df2.columns
```

```python
df2.head()
```

2.5 Finding duplicate values

```python
df2.loc[df2.duplicated()] # No duplicate values found in the dataset
```

```python
df2.loc[df2.duplicated(subset=['Coaster_Name'])].head(5)
```

```python
#checking a duplicate subset value
df2.query('`Coaster_Name` == "Crystal Beach Cyclone"')
```

```python
df2.columns
```

```python
unique_df = df2.drop_duplicates(
    subset=['Coaster_Name', 'Location', 'Opening_Date'],
    keep='first'
).reset_index(drop=True).copy()
```

#### Step 3: Feature Understanding

##### Plotting Feature Distributions

```python
unique_df['Year_Introduced']\
    .value_counts()
    
```

3.1 Bar Chart

```python
# Year Introduced Distribution

ax = unique_df['Year_Introduced']\
    .value_counts()\
    .head(10)\
    .plot(kind='bar', figsize=(12, 6), color='orange', title='Top 10 Years with Most Coasters Introduced')
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Number of Coasters Introduced')
```

3.2 Histogram

```python
# Speed Distribution
ax = unique_df['Speed-Mph']\
    .plot(kind='hist', color='green', bins=20, title='Coasters Speeds Distribution')
ax.set_xlabel('Speed (mph)')
ax.set_ylabel('Number of Coasters')
```

```python
count_40_60 = unique_df[
    (unique_df['Speed-Mph'] >= 40) & 
    (unique_df['Speed-Mph'] <= 60)
].shape[0]

print(f"Number of coasters with speed between 40 and 60 mph: {count_40_60}")
```

```python
count_60_80 = unique_df[
    (unique_df['Speed-Mph'] >= 60) & 
    (unique_df['Speed-Mph'] <= 80)
].shape[0]

print(f"Number of coasters with speed between 60 and 80 mph: {count_60_80}")
```

```python
count_80_100 = unique_df[
    (unique_df['Speed-Mph'] >= 80) & 
    (unique_df['Speed-Mph'] <= 100)
].shape[0]

print(f"Number of coasters with speed between 80 and 100 mph: {count_80_100}")
```

```python
count_100_120 = unique_df[
    (unique_df['Speed-Mph'] >= 100) & 
    (unique_df['Speed-Mph'] <= 120)
].shape[0]

print(f"Number of coasters with speed between 100 and 120 mph: {count_100_120}")
```

```python
count_120_140 = unique_df[
    (unique_df['Speed-Mph'] >= 120) & 
    (unique_df['Speed-Mph'] <= 140)
].shape[0]

print(f"Number of coasters with speed between 120 and 140 mph: {count_120_140}")
```

```python
count_speed_140_more = unique_df[
    (unique_df['Speed-Mph'] > 140)
].shape[0]

print(f"Number of coasters with speed above 140 mph: {count_speed_140_more}")
```

```python
# Manufacturer Distribution
ax = unique_df['Manufacturer']\
    .value_counts()\
    .head(10)\
    .plot(kind='bar', figsize=(12, 6), color='purple', title='Top 10 Manufacturers with Most Coasters')
ax.set_xlabel('Manufacturer')
ax.set_ylabel('Number of Coasters')
```

```python
# opening_date Distribution
ax = unique_df['Opening_Date']\
    .value_counts()\
    .head(10)\
    .plot(kind='bar', figsize=(12, 6), color='blue', title='Top 10 Opening Dates with Most Coasters')
ax.set_xlabel('Opening Date')
ax.set_ylabel('Number of Coasters')
```

3.3 KDE

```python
# Speed Distribution
sns.histplot(data=unique_df, x='Speed-Mph', kde=True, bins=20)
plt.title('Speed Distribution with KDE')
plt.xlabel('Speed (mph)')
plt.show()
```
